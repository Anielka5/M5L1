```python
!pip install rembg, numpy, cv2, google.colab.patches
!pip install rembg[cpu]
from rembg import remove as remove
import numpy as np
import cv2 as image
from google.colab.patches import cv2_imshow as image_show
def remove_and_replace_background(background, url):
  from rembg import remove as remove
  import numpy as np
  import cv2 as image
  from google.colab.patches import cv2_imshow as image_show
  with open(url, 'rb') as i:
    with open('output.png', 'wb') as o:
        input = i.read()
        output = remove(input)
        o.write(output)
  def change(background1, img):
    x, y = 0, 0
    background1 = image.resize(background, (img.shape[1], img.shape[0]), interpolation = image.INTER_AREA)
    res = np.copy(background1)
    place = res[y: y + img.shape[0], x: x + img.shape[1]]
    a = img[..., 3:].repeat(3, axis=2).astype('uint16')
    place[...] = (place.astype('uint16') * (255 - a) // 255) + img[..., :3].astype('uint16') * a // 255

    return res
    with open('1482.png', 'wb') as o:
      o.write(res)





  background = image.imread(background)
  img = image.imread("/content/output.png", image.IMREAD_UNCHANGED)

  result = change(background, img)
  image_show(result)
  image = image.imread(result)
  # The text string
  text = "Mem mojego autorstwa"
  # Czcionka
  font = image.FONT_HERSHEY_SIMPLEX
  # Rozmiar czcionki
  font_scale = 1
  # Kolor w formacie BGR (niebieski, zielony, czerwony).
  # To jest kolor biały
  color = (255, 255, 255)
  # Myślenie o linii
  thickness = 2
  image_with_text = image.putText(image, text, (x, y), font, font_scale, color, thickness, image.LINE_AA)
  image_show(image_with_text)
  image.imwrite(result, image_with_text)


```
